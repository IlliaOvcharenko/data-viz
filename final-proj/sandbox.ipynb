```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime

from pathlib import Path
```

```python
data_folder = Path("data")
```

```python
df = pd.read_csv(
    data_folder / "Apple Music Play Activity.csv", 
    parse_dates=[
        'Event End Timestamp', 
        'Event Received Timestamp',
        'Event Start Timestamp'
    ]
)

df["year_month"] = df["Event Start Timestamp"].apply(lambda dt: f"{dt.year}-{dt.month}")
df["year"] = df["Event Start Timestamp"].apply(lambda dt: dt.year)
df["month"] = df["Event Start Timestamp"].apply(lambda dt: dt.month)
```

```python
# df["Artist Name"].unique()
```

```python
plt.figure(figsize=(20, 7))

for artist_name, artist_df in df.groupby("Artist Name"):
    if artist_df.shape[0] < 300:
        continue
    artist_df = artist_df[artist_df.year_month != "nan-nan"]
    artist_df = artist_df.sort_values(by="Event Start Timestamp")
    grouped = artist_df.groupby("year_month").size()
    artist_df["tmp_col"] = artist_df["yea"]
#     grouped.index, grouped.values
    plt.plot(grouped.index, grouped.values, label=artist_name)
    
plt.legend()
plt.grid(alpha=0.5)
```
